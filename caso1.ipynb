```python
tp52017 = pd.DataFrame()
tp52017['Empresas'] = [df['RAZON SOCIAL'].values[0], df['RAZON SOCIAL'].values[8], df['RAZON SOCIAL'].values[4],
    df['RAZON SOCIAL'].values[31], df['RAZON SOCIAL'].values[11]]
tp52017['Ganacia/Perdida'] = [df['GANANCIA (PERDIDA) 2017'].values[0], df['GANANCIA (PERDIDA) 2017'].values[8], 
    df['GANANCIA (PERDIDA) 2017'].values[4], df['GANANCIA (PERDIDA) 2017'].values[31], 
    df['GANANCIA (PERDIDA) 2017'].values[11]]
```

```python
tp52017
```

```python
import matplotlib.pyplot as plt

plt.figure()    #Figura. Puede incluirse el tamaño con figsize
plt.rcParams["figure.figsize"] = (5, 15)
plt.bar(tp52017['Empresas'], tp52017['Ganacia/Perdida'])          #El gráfico
plt.title('2017')      #El título
ax = plt.subplot()     
ax.set_xlabel('A')  #Nombre del eje x
ax.set_ylabel('B')  #Nombre del eje y

```
